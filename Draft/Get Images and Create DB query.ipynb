```python
for index, row in df.iterrows():
    print(str(row['Short description']))
```

```python
#product
for index, row in df.iterrows():
    query="INSERT INTO tbl_product(id, name, price, category, available) VALUES('"
    query+=str(row['STT'])+"', '"
    query+=row['Name'].strip()+"', '"
    query+=str(row['price'])+"', '"
    if row['Name'].strip()[0]=='B':
        category="Ban phim"
    elif row['Name'].strip()[0]=="C":
        category="Chuot"
    else:
        category="Tai nghe"
    query+=category+"', '"
    query+="1');"
    print(query)
```

```python
#image
for index, row in df.iterrows():
    query="INSERT INTO tbl_image(product_id, path) VALUES('"
    query+=str(row['STT'])+"', '"
    for path in row["Images"]:
        tmp_query=query+"/Product_Images/"+path.split('/')[-1]+"');"
        print(tmp_query)
```

```python
#short description
for index, row in df.iterrows():
    query="INSERT INTO tbl_description(product_id, content) VALUES('"
    query+=str(row['STT'])+"', '"
    for description in row["Short description"]:
        tmp_query=query+description+"');"
        print(tmp_query)
```
